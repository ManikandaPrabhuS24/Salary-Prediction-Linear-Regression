```python
# Model Deployment
import pickle
import pandas as pd
load_model = pickle.load(open("finalized_model.sav",'rb'))
# Default input Prediction
# Default input : 8.7
r=8.7
x = pd.DataFrame([[r]], columns=["YearsExperience"])
prediction = load_model.predict(x)
result = prediction.item()
print(f"The predicted salary for {r} years of experience is {result:2f}")
```

```python
# User input Prediction
r = float(input("Enter the Years Of Experience: "))
inp = pd.DataFrame([[r]], columns=["YearsExperience"])
prediction = load_model.predict(inp)
result = prediction.item()
print(f"The predicted salary for {r} years of experience is {result:2f}")
```
